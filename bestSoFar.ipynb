```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image_path = "/content/zl5956se8had1.jpeg"
image = cv2.imread(image_path)

# Display the original image
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.show()
```

```python
from ultralytics import YOLO

# Load the YOLOv8n model
model = YOLO("/content/bestde.pt")

# Perform detection
results = model(image)

# Extract bounding boxes
bounding_boxes = results[0].boxes.xyxy.cpu().numpy()

# Draw bounding box on the image
for box in bounding_boxes:
    x1, y1, x2, y2 = map(int, box)
    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)

# Display the image with bounding box
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Image with License Plate Detection")
plt.show()
```

```python
# Crop the license plate
for box in bounding_boxes:
    x1, y1, x2, y2 = map(int, box)
    cropped_plate = image[y1:y2, x1:x2]
    cv2.imwrite("cropped_plate.png", cropped_plate)

# Display the cropped license plate
plt.imshow(cv2.cvtColor(cropped_plate, cv2.COLOR_BGR2RGB))
plt.title("Cropped License Plate")
plt.show()
```

```python
# Convert to Grayscale and Apply Gaussian Blur
image = cv2.imread("cropped_plate.png")  # Load the image
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
blurred = cv2.GaussianBlur(gray, (5, 5), 0)  # Apply Gaussian blur

# Show Step 2 output
from google.colab.patches import cv2_imshow
cv2_imshow(blurred)
cv2.waitKey(0)
```

```python
import numpy as np

kernel = np.ones((3, 3), np.uint8)
bLimage = cv2.morphologyEx(blurred, cv2.MORPH_CLOSE, kernel)

#morph = cv2.morphologyEx(bLimage, cv2.MORPH_OPEN, kernel)

from google.colab.patches import cv2_imshow
cv2_imshow(bLimage)
cv2.waitKey(0)
```

```python
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,18))
enhanced_contrast = clahe.apply(bLimage)

from google.colab.patches import cv2_imshow
cv2_imshow(enhanced_contrast)
cv2.waitKey(0)
```

```python
contrast_image = cv2.equalizeHist(enhanced_contrast)

from google.colab.patches import cv2_imshow
cv2_imshow(contrast_image)
cv2.waitKey(0)
```

```python
#Thresholding using Otsu's method
_, thresholded = cv2.threshold(enhanced_contrast, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Invert the thresholded image to make the plate black and the text white
thresholded_inverted = cv2.bitwise_not(thresholded)


from google.colab.patches import cv2_imshow
cv2_imshow(thresholded_inverted)
cv2.waitKey(0)
```

```python
# Step 4: Dilate the image to make contours more visible
dilated = cv2.dilate(thresholded_inverted, None, iterations=1)


from google.colab.patches import cv2_imshow
cv2_imshow(dilated)
cv2.waitKey(0)
```

```python
contours, _ = cv2.findContours(dilated, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)


rectangular_contours = []
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)


    if 8 < w < 130 and 15 < h < 200:
        rectangular_contours.append((x, y, w, h))


rectangular_contours = sorted(rectangular_contours, key=lambda c: (c[1] // 10, c[0]))


image_with_boxes = image.copy()
for (x, y, w, h) in rectangular_contours:
    cv2.rectangle(image_with_boxes, (x, y), (x + w, y + h), (0, 255, 0), 2)


from google.colab.patches import cv2_imshow
cv2_imshow(image_with_boxes)
cv2.waitKey(0)
```

```python
import easyocr
import cv2


filtered_contours = []
valid_characters = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"


reader = easyocr.Reader(['en'])

for (x, y, w, h) in rectangular_contours:
    if w > 10 and h > 20:

        char_img = thresholded_inverted[y:y + h, x:x + w]


        result = reader.readtext(char_img, paragraph=False)

        for detection in result:
            text = detection[1].strip()


            if text and all(c in valid_characters for c in text.upper()):
                filtered_contours.append((x, y, w, h))


image_filtered_boxes = image.copy()
for (x, y, w, h) in filtered_contours:
    cv2.rectangle(image_filtered_boxes, (x, y), (x + w, y + h), (0, 255, 0), 2)


from google.colab.patches import cv2_imshow
cv2_imshow(image_filtered_boxes)
cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec


crop_characters = []


for (x, y, w, h) in filtered_contours:
    char_img = thresholded_inverted[y:y + h, x:x + w]
    crop_characters.append(char_img)


fig = plt.figure(figsize=(14, 4))
grid = gridspec.GridSpec(ncols=len(crop_characters), nrows=1, figure=fig)

for i in range(len(crop_characters)):
    fig.add_subplot(grid[i])
    plt.axis(False)
    plt.imshow(crop_characters[i], cmap="gray")


# plt.savefig("segmented_letter.png", dpi=300)

plt.show()
```

```python
# Step 7: Segment each character and apply bitwise_not
character_images = []
for (x, y, w, h) in filtered_contours:
    # Extract character region
    char_img = thresholded_inverted[y:y + h, x:x + w]

    # Apply bitwise_not to invert the image for Tesseract
    char_img_inverted = cv2.bitwise_not(char_img)

    # Apply median blur for better Tesseract accuracy
    char_img_blurred = cv2.medianBlur(char_img_inverted, 3)

    # Append character images for OCR
    character_images.append(char_img_blurred)

    # Draw bounding box for each character
    cv2.rectangle(image_filtered_boxes, (x, y), (x + w, y + h), (0, 0, 255), 2)

# Show Step 7 output
cv2_imshow(image_filtered_boxes)
cv2.waitKey(0)
```

```python
import easyocr


reader = easyocr.Reader(['en'])

license_plate = ""


for char_img in crop_characters:
    result = reader.readtext(char_img)
    if result:
        text = result[0][1]
        license_plate += text.strip()

print("License Plate Recognized:", license_plate)
```
